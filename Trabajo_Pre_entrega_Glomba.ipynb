##06. Limpieza de datos

```python
#Limpiamos y normalizamos los Dataframes:
#Ventas, Marketing, y Clientes

#Creamos nuevos destino en dataframe para no afectar los actuales

df_ventas_clean = df_ventas.copy()
df_marketing_clean = df_marketing.copy()
df_clientes_clean = df_clientes.copy()

#Eliminamos las filas duplicadas
df_ventas_clean = df_ventas_clean.drop_duplicates()
df_marketing_clean = df_marketing_clean.drop_duplicates()
df_clientes_clean = df_clientes_clean.drop_duplicates()
```

```python
#Funcion para limpiar texto en columnas tipo string

#Python nos va a ayudar a analizar el bloque
def normalizar_texto(df):
  for col in df.select_dtypes(include="object"):
    df_transformado = df[col].astype(str) # Convierte cualquier tipo a string
    df_transformado = df_transformado.str.strip() # Elimina espacios al inicio y final
    df_transformado = df_transformado.str.replace(r"[\u200b\t\r\n\xa0]", "", regex=True)
        # .str.replace(patron, reemplazo, regex=True)
            #   patron: expresi칩n regular que busca caracteres invisibles (\u200b, tabulaciones, saltos)
            #   reemplazo: ""   los elimina
            #   regex=True indica que 'patron' es una expresi칩n regular.
    df_transformado = df_transformado.str.replace(" +"," ", regex=True)
    df_transformado = df_transformado.str.title()
    df[col] = df_transformado
  return df
```

```python
#Normalizamos las fechas

#Intentamos convertir a formato datetime de pandas, mediante to_datetime, alguna columna que contenga fechas

for df in [df_ventas_clean, df_clientes_clean, df_marketing_clean]:
    for col in df.columns:
        if "fecha" in col.lower():  # detecta columnas con la palabra "fecha"
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
            # errors="coerce" } convierte valores no v치lidos en NaT (evita error)
            #   dayfirst=True    interpreta formatos tipo "DD/MM/YYYY" (formato latino)
```

```python
#Normalizamos valores numericos
#Campo "precio" del dataframe de ventas para modificar los datos en int

df_ventas_clean["precio"] = df_ventas_clean["precio"].str.replace("$", "", regex=False)
```

```python
#Se aplica la normalizacion del texto
df_ventas_clean = normalizar_texto(df_ventas_clean)
df_clientes_clean = normalizar_texto(df_clientes_clean)
df_marketing_clean = normalizar_texto(df_marketing_clean)
```
